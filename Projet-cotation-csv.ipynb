```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from numpy import genfromtxt
```

```python
columns = ['date', 'value', 'volume']
```

```python
data_kering = pd.read_csv('KERING_bourse.csv',sep=';', names = columns, skiprows=1)
data_hermes = pd.read_csv('HERMES_bourse.csv',sep=';', names = columns, skiprows=1)
data_lvmh = pd.read_csv('LVMH_bourse.csv',sep=';', names = columns, skiprows=1)
data_CAC40 = pd.read_csv('CAC40.csv',sep=';', names = columns, skiprows=1)
```

```python
df_data_kering = pd.DataFrame(data_kering)
df_data_hermes = pd.DataFrame(data_hermes)
df_data_lvmh = pd.DataFrame(data_lvmh)
df_data_cac40 = pd.DataFrame(data_CAC40)
```

```python
# ne pas utiliser
import matplotlib.pyplot as plt
import csv
  
x = []
y = []
  
with open('HERMES_bourse.csv','r') as csvfile:
    plots = csv.reader(csvfile, delimiter = ';')
      
    for row in plots:
        x.append(row[0])
        y.append(row[1])
  
plt.bar(x, y, label = "Date")
plt.xlabel('Date')
plt.ylabel('Value')
plt.title('Actions value')
plt.legend()
plt.show()
```

```python
import pytrends 
# import the TrendReq method from the pytrends request module
from pytrends.request import TrendReq

# execute the TrendReq method by passing the host language (hl) and timezone (tz) parameters
pytrends = TrendReq(hl='fr', tz=360)
```

```python
# build list of keywords
kw_list = ["Louis Vuitton", "Gucci", "Hermes"] 

# build the payload
pytrends.build_payload(kw_list, timeframe='2020-01-01 2021-01-01', geo='FR')
```

```python
df = pytrends.interest_over_time()
print(df.head(20))
```

```python
plt.figure()
plt.plot(df.index,df.data_kering,'k*')
plt.plot(df.index,df.data_hermes,'r*')
plt.plot(df.index,df.data_kering,'b*')
plt.legend(['Louis Vuitton','Gucci','Hermes'])
```
